##### ¿Puedes hacer que devuelva dos decimales? ¿Y que lidie con mayúsculas y minúsculas?

```python
def calculate_a_percentage(dna_sequence):
    """
    Calculates the percentage of 'A's in a DNA sequence, case-insensitive.

    Args:
        dna_sequence: A string representing the DNA sequence.

    Returns:
        The percentage of 'A's in the sequence, formatted to two decimal places.
    """
    a_count = 0
    for nucleotide in dna_sequence:
        if nucleotide.lower() == 'a':
            a_count += 1
    total_length = len(dna_sequence)
    if total_length == 0:
        return 0.00
    else:
        return round((a_count / total_length) * 100, 2)
dna = "aTcGatCg"
percentage = calculate_a_percentage(dna)
print(f"The percentage of 'A's in '{dna}' is: {percentage}%")
```

#### 5.- De la lista de accessions que se proporciona obtén lo siguiente:
1. Los que tengan el número 5
2. Los que tengan la letra 'd' o la 'e'
3. Los que tengan las letras 'd' y 'e' en ese orden
4. Los que tengan las letras 'd' y 'e' en ese orden con una letra cualquiera en medio
5. Los que tengan ambas letras 'd' y 'e' en cualquier orden
6. Los que empiezan por 'x' o por 'y'
7. Los que empiezan por 'x' o 'y' y terminan por 'e'
8. Los que tienen tres o más dígitos seguidos
9. Los que terminan con 'd' seguidos de 'a' o 'r' o 'p'

```python
import re

accs = ['xkn59438', 'yhdck2', 'eihd39d9', 'chdsye847', 'hedle3455', 'xjhd53e', '45da', 'de37dp']

print("Accessions with the number 5:")
for acc in accs:
    if re.search(r'5', acc):
        print(acc)

print("\nAccessions with the letter 'd' or 'e':")
for acc in accs:
    if re.search(r'[de]', acc):
        print(acc)

print("\nAccessions with 'd' and 'e' in that order:")
for acc in accs:
    if re.search(r'd.*e', acc):
        print(acc)

print("\nAccessions with 'd' and 'e' in that order with one character in between:")
for acc in accs:
    if re.search(r'd.e', acc):
        print(acc)

print("\nAccessions with both 'd' and 'e' in any order:")
for acc in accs:
    if re.search(r'(?=.*d)(?=.*e)', acc):
        print(acc)

print("\nAccessions starting with 'x' or 'y':")
for acc in accs:
    if re.search(r'^[xy]', acc):
        print(acc)

print("\nAccessions starting with 'x' or 'y' and ending with 'e':")
for acc in accs:
    if re.search(r'^[xy].*e$', acc):
        print(acc)

print("\nAccessions with three or more consecutive digits:")
for acc in accs:
    if re.search(r'\d{3,}', acc):
        print(acc)

print("\nAccessions ending with 'd' followed by 'a', 'r', or 'p':")
for acc in accs:
    if re.search(r'd[arp]$', acc):
        print(acc)
```
